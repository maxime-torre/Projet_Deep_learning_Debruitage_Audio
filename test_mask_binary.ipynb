```python
from pathlib import Path
from mask_binary import  calculate_amplitude_spectrogram, mask_binary
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import os
```

```python
N = 2048
H = 1024
```

```python
noise_audio_path =  Path.cwd() / "data" / "only_noise" / "20.flac"# Remplacer par le chemin r√©el
audio_raw_data_path = Path.cwd() / "data" / "raw_data_cut" / "20.flac" 
soft_mask = mask_binary(audio_raw_data_path, noise_audio_path, N, H, mask_type='binary')
signal_and_noise = Path.cwd() / "data" / "noised_data" / "10" / "20.flac"
```

```python
#load the signal + noise
y_signal_noise, sr = librosa.load(signal_and_noise)
display(Audio(y_signal_noise, rate=sr))
```

```python
#load the noise
y_noise, sr = librosa.load(noise_audio_path)
display(Audio(y_noise, rate=sr))
```

```python
#calcul du spectrogramme du signal bruit√©
spectro_signal_noise = calculate_amplitude_spectrogram(signal_and_noise, N , H)

print(f"len(spectro_signal_noise) : {len(spectro_signal_noise)}")
print(f"len(soft_mask) : {len(soft_mask)}")

#Application du masque
signal_and_noise_with_mask = spectro_signal_noise*soft_mask

#Retour au signal temporel avec l'ISTFT
reconstructed_signal = librosa.istft(signal_and_noise_with_mask, hop_length=H, win_length=N, window='hann')

display(Audio(reconstructed_signal, rate=sr))
```
